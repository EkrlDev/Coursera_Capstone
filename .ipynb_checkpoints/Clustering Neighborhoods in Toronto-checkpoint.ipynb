# Part-1 Preparing Data Frame

```python
# Ä°mporting dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
#Scraping the WEB Site
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
source = requests.get(url).text

sourceXML = BeautifulSoup(source, 'lxml')
table=sourceXML.find('table')

#creating dataframe with the columns: PostalCode, Borough, and Neighborhood
column_names = ['Postalcode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)

tableRows = table.find_all('tr')

for tr in tableRows:
    row=[]
    for td in tr.find_all('td'):
        row.append(td.text.strip())
    if len(row)==3:
        df.loc[len(df)] = row

#Cleanin Not indexed rows
index = df[df['Borough'] == 'Not assigned'].index
df.drop(index, inplace = True)
        
df.head(5)
```

```python
df.shape
```

# Part-2 Adding Geographic Locations

```python
df_coor = pd.read_csv("https://cocl.us/Geospatial_data")

df_coor.rename(columns={'Postal Code':'Postalcode'},inplace=True)
merged_df = pd.merge(df, df_coor, on='Postalcode')

merged_df.head()
```

# Part-3 Clustering Neighborhoods

```python
!pip install folium
import folium # plotting library
```

```python
from geopy.geocoders import Nominatim
```

```python
address = "Toronto, ON"

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto city are {}, {}.'.format(latitude, longitude))

toronto_map = folium.Map(location=[latitude, longitude], zoom_start=13)

#Adding markers on map
for lat, lng, borough, neighborhood in zip(
        merged_df['Latitude'], 
        merged_df['Longitude'], 
        merged_df['Borough'], 
        merged_df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='Blue',
        fill=True,
        fill_color='Green',
        fill_opacity=0.7,
        parse_html=False).add_to(toronto_map)  


toronto_map
```

```python
toronto_nh = merged_df.groupby('Neighborhood')
```

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
nh_grouped = merged_df.groupby('Neighborhood').mean().reset_index()
nh_grouped.head()
```

```python
# set number of clusters
kclusters = 5

toronto_nh_clustering = nh_grouped.drop('Neighborhood', 1)
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_nh_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
nh_grouped.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
nh_grouped
```

```python


# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(nh_grouped['Latitude'], nh_grouped['Longitude'], nh_grouped['Neighborhood'], nh_grouped['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(toronto_map)
toronto_map
```
